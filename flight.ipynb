```python
def time(x):
    if x>=1 and x<=400:
        return 'Mid_n'
    elif x>400 and x<=800:
        return 'early_m'
    elif x>800 and x<=1200:
        return 'morning'
    elif x>1200 and x<=1600:
        return 'afternoon'
    elif x>1600 and x<=2000:
        return 'evening'
    else:
        return 'night'
    
```

```python
df['Category']=df.SCHED_DEP.apply(time)
df
```

```python
df['price']=df.DEP_DELAY.apply(time)
df
```
